```python
df['Postal Code'] = df['Postal Code'].astype('object')
```

```python
df.drop_duplicates(subset=None,keep='first',inplace=True) #delete duplicates
df.duplicated().sum()     
```

```python
corr = df.corr()
sns.heatmap(corr,annot=True,cmap='Greens')
```

```python
#dropping Postal Code column
df = df.drop(['Postal Code'],axis=1)  
```

```python
sns.pairplot(df, hue = 'Region')   #plotting pair plot
```

```python
df.Region.unique()
```

```python
df.Region.value_counts()
```

Distribution of Data -Regionwise

```python
regions = ['Central', 'East','South','West']
count = [2319, 2845, 1620, 3193]

plt.figure(figsize=(14,6))
plt.subplot(1,2,1)

#barplot
plt.bar(regions, count, color='red', ec='black')
plt.title("Distribution - Regions")
plt.xlabel("Regions")
plt.ylabel("Count")
plt.show()
```

```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['darkblue','skyblue'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

profit and sales are maximum in west region and minimum in south region. Central and south data as shows.

```python
fig,ax=plt.subplots(figsize=(20,8))
ax.scatter(df['Sales'],df['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=df)
plt.legend()
plt.show()
```

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['green','yellow'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

Profit and sales are maximum in consumer segment and minimum in Home Office segment

```python
ps = df.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['pink','blue'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

high profit is for california, new york
loss is for texas, pennsylvania, Ohio

```python
large_to_small_states = df['State'].value_counts().nlargest(10)
large_to_small_states
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['pink','red'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.set(font_scale=1.4)
sns.barplot(x='Sub-Category', y='Profit', data=df, palette= "bright")
plt.title('Profit Graph for Technology all over USA')
plt.show()
```

As a business manager, try to find out the weak areas where you can work to make more profit?



```python
ps = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['purple','yellow'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

As conclusion, Phones sub-category have high sales.Also, chairs have high sales but less profit compared to phones.

Tables and Bookcases sub-categories facing huge loss.And others sub-categories as shows for extracting knowledge to predict decision.
